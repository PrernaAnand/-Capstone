```python
from google.colab import drive
path = drive.mount('/content/drive/')
import collections
all_data = collections.defaultdict(list)
```

```python
import csv

def data_cleanup(file_name):
  with open(file_name, newline='') as csvfile:
    content = csv.reader(csvfile)
    line_count = 0
    for row in content:
      if line_count > 0:
        X = float(row[0])
        Y = float(row[1])
        Power = float(row[3])
        all_data[(X,Y)].append(Power)
      
      line_count += 1  
```

```python
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/Convergent analysis/one.csv')
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/Convergent analysis/two.csv')
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/Convergent analysis/three.csv')
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/Convergent analysis/four.csv')
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/Convergent analysis/five.csv')
```

```python
import numpy as np

arr1 = []
arr2 = []
arr3 = []
arr4 = []
arr5 = []
for _, values in all_data.items():
  arr1.append(values[0])
  arr2.append(values[1])
  arr3.append(values[2])
  arr4.append(values[3])
  arr5.append(values[4])

arr1 = np.array(arr1)
arr2 = np.array(arr2)
arr3 = np.array(arr3)
arr4 = np.array(arr4)
arr5 = np.array(arr5)

arrays = [arr1,arr2,arr3,arr4,arr5]
```

```python
result = float('inf')
convergent_index = None

i = 0
while i+1 < len(arrays):
    diff = arrays[i+1] - arrays[i]
    res = sum(diff)/len(diff)
    if res < result:
      result = res
      convergent_index = i+2
    i += 1

print(convergent_index)
```
