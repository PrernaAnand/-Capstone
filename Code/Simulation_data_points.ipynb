```python
import numpy as np
import pandas as pd
import os
from google.colab import drive

results = []
```

```python
# Data Points

def find_data_points(rectangle_bounds, results):
  for i in np.arange(rectangle_bounds[0][0], rectangle_bounds[0][1], 0.5):
    for j in np.arange(rectangle_bounds[1][0], rectangle_bounds[1][1], 0.5):
      results.append([i,j,0])
      
def export_file(arr):
  dir = os.getcwd()
  path = drive.mount('/content/drive/')
  np.savetxt(r'/content/drive/MyDrive/Capstone Material/Code/results.txt', arr, fmt='%f')
```
