```python
import collections
all_data = collections.defaultdict(list)
test_data = collections.defaultdict(list)
```

```python
import csv

def data_cleanup(file_name):
  with open(file_name, newline='') as csvfile:
    content = csv.reader(csvfile)
    line_count = 0
    for row in content:
      if line_count > 0:
        X = float(row[0])
        Y = float(row[1])
        Power = float(row[3])
        all_data[(X,Y)].append(Power)
      
      line_count += 1  
```

```python
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/source1.csv')
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/source2.csv')
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/source3.csv')
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/source4.csv')
data_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/source5.csv')

result = [["X", "Y", "S1", "S2", "S3", "S4", "S5"]]
for key, values in all_data.items():
  result.append([key[0], key[1], values[0], values[1], values[2], values[3], values[4]])
print(result)

with open(r'/content/drive/MyDrive/Capstone Material/Code/final_train_data.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(result)
```

```python
def data_test_cleanup(file_name):
  with open(file_name, newline='') as csvfile:
    content = csv.reader(csvfile)
    line_count = 0
    for row in content:
      if line_count > 0:
        X = float(row[0])
        Y = float(row[1])

        if (X,Y) not in all_data:
          Power = float(row[3])
          test_data[(X,Y)].append(Power)
      
      line_count += 1  
```

```python
data_test_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/testing_source1.csv')
data_test_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/testing_source2.csv')
data_test_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/testing_source3.csv')
data_test_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/testing_source4.csv')
data_test_cleanup(r'/content/drive/MyDrive/Capstone Material/Code/testing_source5.csv')

result = [["X", "Y", "S1", "S2", "S3", "S4", "S5"]]
for key, values in test_data.items():
  result.append([key[0], key[1], values[0], values[1], values[2], values[3], values[4]])
print(result)

with open(r'/content/drive/MyDrive/Capstone Material/Code/final_test_data.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(result)
```
