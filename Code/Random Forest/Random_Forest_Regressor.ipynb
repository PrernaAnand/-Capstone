```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Capstone Material/Code/final_train_data.csv')
X = data.iloc[:,2:7]
y = data.iloc[:,0:2]

data_test = pd.read_csv('/content/drive/MyDrive/Capstone Material/Code/final_test_data.csv')
X_test = np.array(data_test.iloc[:,2:7])
y_test = np.array(data_test.iloc[:,0:2])

regressor = RandomForestRegressor(n_estimators=2000, random_state=0, max_depth=100, min_samples_split=2,min_samples_leaf=1,max_features='auto',bootstrap=True)
regressor.fit(X, y)
y_pred = regressor.predict(X_test)
metrics.mean_absolute_error(y_test, y_pred)
```

```python
y_pred_depth = list()
error = list()
for i in range(len(X_test)):
  print(i)
  y_pred = regressor.predict([X_test[i]])[0]
  error.append(metrics.mean_absolute_error(y_test[i], y_pred))
  y_pred_depth.append((y_test[i], y_pred))
print(y_pred_depth)
```

```python
import matplotlib.pyplot as plt

x = range(100,200)
y = range(100,200)
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter([i[0][0] for i in y_pred_depth], [i[0][1] for i in y_pred_depth], s=1, c='b', marker="s", label='Actual value')
ax1.scatter([i[1][0] for i in y_pred_depth], [i[1][1] for i in y_pred_depth], s=1, c='r', marker="o", label='Predicted value')
plt.legend(loc="upper right")
plt.show()
```

```python
from sklearn.preprocessing import normalize
norm_errors = normalize(np.expand_dims(error, 1), axis=0)
norm_errors = np.squeeze(norm_errors, 1)
```

```python
error_count = {}
for i in norm_errors:
  if i not in error_count:
      error_count[i] = 0
  error_count[i] += 1
```
